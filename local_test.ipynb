```python
import requests
import json
```

```python
def get_completion(prompt):
    headers = {'Content-Type': 'application/json'}
    data = {"prompt": prompt}
    response = requests.post(url='http://162.105.88.63:6006', headers=headers, data=json.dumps(data))
    return response.json()['response']
```

```python
print(get_completion('你是谁，请给我介绍下自己'))
```
